```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'RELIANCE.csv'
data = pd.read_csv(file_path)
```

```python
data
```

```python
# Data Cleaning
# Convert Date to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Handle missing values
data['Trades'].fillna(0, inplace=True)
data['Deliverable Volume'].fillna(0, inplace=True)
data['%Deliverble'].fillna(0, inplace=True)
```

```python
# Feature Engineering
# Calculate daily price range
data['Price Range'] = data['High'] - data['Low']
# Calculate daily price change percentage
data['Price Change %'] = ((data['Close'] - data['Open']) / data['Open']) * 100
# Calculate turnover per trade
data['Turnover per Trade'] = np.where(data['Trades'] > 0, data['Turnover'] / data['Trades'], 0)
```

```python
# Statistical Summary
# Generate descriptive statistics
descriptive_stats = data.describe()
print("Descriptive Statistics:")
print(descriptive_stats)
```

```python
# Anomaly Detection
# Select relevant features for analysis
features = ['Price Range', 'Volume', 'Turnover', 'VWAP', 'Price Change %', 'Turnover per Trade']
scaler = StandardScaler()
X = scaler.fit_transform(data[features])

# Apply Isolation Forest for anomaly detection
iso_forest = IsolationForest(contamination=0.01, random_state=42)
data['Anomaly'] = iso_forest.fit_predict(X)

# Mark anomalies (1 for normal, -1 for anomaly)
data['Anomaly'] = data['Anomaly'].apply(lambda x: 1 if x == -1 else 0)
```

```python
# Anomaly Analysis
# Count anomalies per year
data['Year'] = data['Date'].dt.year
anomalies_per_year = data[data['Anomaly'] == 1].groupby('Year').size()
print("Anomalies Detected Per Year:")
print(anomalies_per_year)

# Visualization
plt.figure(figsize=(12, 6))
plt.title('Anomaly Detection in Price Range and Volume')
sns.scatterplot(
    x=data['Date'],
    y=data['Price Range'],
    hue=data['Anomaly'],
    palette={0: 'blue', 1: 'red'},
    legend='full'
)
plt.xlabel('Date')
plt.ylabel('Price Range')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Anomalies Over Time')
sns.lineplot(data=data, x='Date', y='Price Change %', hue='Anomaly', palette={0: 'green', 1: 'orange'})
plt.xlabel('Date')
plt.ylabel('Price Change %')
plt.show()
```

```python
data.to_csv('RELIANCE_analyzed.csv', index=False)
```
